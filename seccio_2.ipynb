```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

Líquid:

```python
# Definim paràmetres inicials amb N=500
T_l_0 = 1.0
den_l_0 = 0.7
# Llista de P_0 per graficar més endavant
P_l_0 = [0.025,0.05,0.075,0.1,0.125]
#Apuntem els resultats obtinguts en la simulació
P_l = np.array([0.021,0.051,0.075,0.098,0.126])
V_l = np.array([714.30,708.66,704.73,701.53,697.19])
lnV_l = np.log(V_l)
print(lnV_l)
```

```python
# Regresión lineal
result = linregress(P_l, lnV_l)

slope = result.slope
intercept = result.intercept
r_value = result.rvalue
r_squared = r_value**2

# Valores ajustados
fit = slope * P_l + intercept

# Gráfica
plt.figure(figsize=(6,4))
plt.scatter(P_l, lnV_l, label='Dades')
plt.plot(P_l, fit, label='Regressió lineal')

plt.xlabel('P')
plt.ylabel('ln(V)')
plt.title('ln(V) en funció de P en liquid')
plt.legend()
plt.grid(True)

plt.show()

# Resultados
print(f"Slope = {slope}")
print(f"Intercept = {intercept}")
print(f"R^2 = {r_squared}")
```
